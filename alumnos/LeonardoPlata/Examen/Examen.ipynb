```python
import pandas as pd
import matplotlib.pyplot as plot
from math import log
```

```python
data = pd.read_csv("crime_data.csv",usecols=['crime','date','hour','lat','long'])
data = pd.DataFrame(data=data)
data.head()
```

```python
fig, cr = plot.subplots()
cr.scatter(data['long'],data['lat'])
plot.xlabel('Longitud')
plot.ylabel('Latitud')
plot.show()
```

```python
fig, cr = plot.subplots()
cr.scatter(data['hour'],data['lat'])
plot.ylabel('Latitud')
plot.xlabel('Tiempo')
plot.show()
```

```python
x = data['long']
y = data['lat']
t = data['hour']
```

```python
n = 8000
def phi(x=x,t=t):
    p = [None]*n
    for i in range(n):
        p[i] = x[i] + x[i+1]*log(abs(int(t[i][0:2]))) - log(abs(x[i+2]*int(t[i][0:2])))
    return p
```

```python
def f(y=y,x=x,t=t):
    r = [None]*n
    p = phi(x,t)
    for i in range(n):
        r[i] = (y[i]-p[i])**2
    return '%.4f' %sum(r) 
```
